```python
x = df.iloc[:,0:12].values
x
```

```python
y = df.iloc[:,13:].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test =train_test_split(x,y, test_size=0.80, random_state=0)
```

```python
x_train.shape,x_test.shape, y_train.shape, y_test.shape
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
regressor = LogisticRegression()
regressor.fit(x_train, y_train)
```

```python
y_predict = regressor.predict(x_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix, precision_score, f1_score, roc_auc_score, precision_recall_fscore_support,recall_score,accuracy_score
```

```python
pd.DataFrame(data=[accuracy_score(y_test, y_predict)*100, recall_score(y_test, y_predict)*100,
                   precision_score(y_test, y_predict)*100, roc_auc_score(y_test, y_predict)*100], 
                    index=["accuracy", "recall", "precision", "roc_auc_score"])
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion='gini',
    splitter='best',
    max_depth=2,
    min_samples_split=20,
    min_samples_leaf=20,
    min_weight_fraction_leaf=0.0,
    max_features=None,
    random_state=None,
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    class_weight=None,
    ccp_alpha=0.0)
```

```python
classifier.fit(x_train, y_train)
```

```python
y_predict = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, roc_auc_score, roc_curve
```

```python
score = pd.DataFrame(data = [confusion_matrix(y_test, y_predict), accuracy_score(y_test, y_predict)*100, 
                             precision_score(y_test, y_predict)*100, roc_auc_score(y_test, y_predict)*100, 
                             roc_curve(y_test, y_predict)], index = ['confusion_matrix', 'accuracy_score', 
                                                                     'precision_score', 'roc_auc_score',
                                                                     'roc_curve'])
```

```python
score
```
